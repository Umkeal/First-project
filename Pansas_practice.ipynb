```python
import pandas as pd # Первая теория # Импортирует библиотеку pandas под коротким именем pd

data = { #Источник данных/словарь на который я опираюсь, который содержит 4 списка
    'id': [1, 2, 3, 4], # Уникальный номер аудитории
    'room_number': [101, 205, 310, 402], # Номер комнаты
    'floor': [1, 2, 3, 4], # Этаж
    'building': ['A', 'B', 'A', 'C'] # Корпус (здание)
}

df = pd.DataFrame(data) # df(DataFrame), здесь я создаю таблицу
print("Таблица:")
print(df) 

print("\nКоличество аудиторий по корпусам:") #\n — это символ переноса строки (newline) в Python.
print(df.groupby('building').size()) # df.groupby('XXXXX') — группирует строки таблицы по уникальным значениям в столбце.
```

```python
# Теория фильтрации, аналог команды WHERE(SQL)
print("=== ВСЕ АУДИТОРИИ ===")
print(df)

print("\n=== 1. ЭТАЖ > 2 (как WHERE floor > 2) ===")
high_floors = df[df['floor'] > 2]   
print(high_floors)

print("\n=== 2. КОРПУС 'A' (как WHERE building = 'A') ===")
building_a = df[df['building'] == 'A']
print(building_a)

print("\n=== 3. 1-й ЭТАЖ И КОРПУС B (как WHERE floor=1 AND building='B') ===")
first_b = df[(df['floor'] == 1) & (df['building'] == 'B')]
print(first_b)
```

```python
import pandas as pd # Практика команды WHERE
data = {
    'id': [1, 2, 3, 4, 5, 6],
    'room_number': [101, 205, 310, 402, 115, 220],
    'floor': [1, 2, 3, 4, 1, 2],
    'building': ['A', 'B', 'A', 'C', 'A', 'B']
}
df = pd.DataFrame(data)

print('Аудитория 2-ого этажа') 
floor_2_au = df[df['floor'] == 2]
print(floor_2_au)

print('Корпус C')
corp_C = df[df['building'] == 'C']
print(corp_C)

print('Комнаты 100+ НОМЕР') 
number_200 = df[df['room_number'] > 200]
print(number_200)

```

```python
print("=== 1. СКОЛЬКО АУДИТОРИЙ ПО КОРПУСАМ ===") # Теория GROUP BY(SQL)


by_building = df.groupby('building').size()
print(by_building)

print("\n=== 2. СКОЛЬКО ПО ЭТАЖАМ ===")
by_floor = df.groupby('floor').size()
print(by_floor)

print("\n=== 3. СРЕДНИЙ НОМЕР КОМНАТЫ ПО КОРПУСУ ===")
avg_room = df.groupby('building')['room_number'].mean()
print(avg_room)

print("\n=== 4. КОЛИЧЕСТВО + СРЕДНИЙ ЭТАЖ ПО КОРПУСУ ===")
summary = df.groupby('building').agg({
    'id': 'count',           # количество строк
    'floor': 'mean'          # средний этаж
})
print(summary)
```

```python
 # Практика команды GROUP BY(SQL)
# 1. Сколько аудиторий по этажам? 
print('\nКоличество аудиторий по этажам:')
by_building = df.groupby('building').size()
print(by_building) 

# 2. Средний этаж по корпусам
print("\nСредний этаж по корпусам:")
avg_floor = df.groupby('building')['floor'].mean()
print(avg_floor)

# 3. Максимальный номер комнаты по корпусу B
print("\nМакс номер в корпусе B:")
max_b = df[df['building'] == 'B']['room_number'].max()
print(max_b) 
```

```python
# Сохранение отчёта в EXCEL
summary = df.groupby('building').agg({
    'id': 'count',        # количество
    'floor': 'mean',      # средний этаж
    'room_number': 'max'  # максимальный номер
}).round(1)

print("Полный отчёт:")
print(summary)

summary.to_csv('classrooms_report.csv')
print("CSV сохранён!")
```
